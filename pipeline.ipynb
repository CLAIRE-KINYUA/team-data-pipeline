```python
import numpy as np
import pandas as pd

data = {
    'Product': ['Laptop', 'Mouse', 'Monitor', 'Keyboard', 'Headphones'],
    'Units_Sold': [15, 120, 30, 75, 50],
    'Unit_Price': [1200, 25, 250, 45, 80]
}

df = pd.DataFrame(data)
df['Total_Revenue'] = df['Units_Sold'] * df['Unit_Price']
df
```
